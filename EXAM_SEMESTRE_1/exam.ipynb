# Problème 1: Analyse et transformation d'un texte
## problem1_text_analysis.py

```python
phrase=input("Entrer une phrase :")
phrase=phrase.casefold()
mots=phrase.split()
print(f"le nombre total de mots est : {len(mots)}")

maxMot=""
for mot in mots :
    if (len(mot)>len(maxMot)):
        maxMot=mot
print(f"le mot le plus long est : {maxMot}")

def totalVoyelles(phrase) :
    voyelles="aeyuio"
    cpt=0
    for lettre in phrase :
        if lettre in voyelles : 
            cpt=cpt+1
    return cpt

print(f"le nombre total de voyelles est : {totalVoyelles(phrase)}")

motsConvertis=[]
for mot in mots :
    if (len(mot) % 2 == 0) :
        motsConvertis.append(mot.upper())
    else :
            motsConvertis.append(mot)
motsConvertis= " ".join(motsConvertis)

print(f'La phrase convertie est : "{motsConvertis}"')
```

# Problème 2 : Gestion des notes d'une classe
## problem2_grades.py

```python
students=[
    ("Ali", 12),
    ("Fatou", 17),
    ("Moussa", 9),
    ("Awa", 14),
    ("Ibrahima", 7)
]

for nom,note in students :
    print(f"{nom} à {note}")

somme = 0
for nom, note in students:
    somme = somme + note
moyenne = somme / len(students)

print(f"\nLmoyenne de la classe est de : {moyenne}\n")

maxNote=students[0][1]
for nom,note in students :
    if note>maxNote :
        maxNote=note
print(f"\nLa meilleur note est {maxNote}")

minNote=students[0][1]
for nom,note in students :
    if note<minNote :
        minNote=note
print(f"La pire note est {minNote}\n")

admis=[]
admis = []
for nom, note in students:
    if note >= 10:
        admis.append(nom)
print("Sont admis :\n")
for nom in admis:
    print(f"{nom}")

ajournee=[]
ajournee = []
for nom, note in students:
    if note < 10:
        ajournee.append(nom)
print("Sont ajournée :\n")
for nom in ajournee:
    print(f"{nom}")

```
